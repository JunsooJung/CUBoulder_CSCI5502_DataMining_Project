```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
games_data = pd.read_csv("Games_Dataset.csv")
```

```python
games_data.head(10)
```

```python
games_data.isna().sum()
```

```python
# dropping the column "playtime_total" since there are a lot of null values
games_data = games_data.drop("playtime_total", axis=1)
games_data
```

```python
games_data.isna().sum()
```

```python
# dropping the rows with null values while adjusting the indexes
games_data = games_data.dropna().reset_index(drop=True)
```

```python
games_data.isna().sum()
```

```python
games_data["Name"] = games_data["Name"].str.replace("[", "")
```

```python
games_data["Name"] = games_data["Name"].str.replace("]", "")
```

```python
games_data["Name"] = games_data["Name"].str.replace('"', "")
```

```python
games_data["Name"] = np.where(games_data["Name"].str.startswith("'"), games_data["Name"].str[1:], games_data["Name"])
games_data["Name"] = np.where(games_data["Name"].str.endswith("'"), games_data["Name"].str[:-1], games_data["Name"])
```

```python
games_data["Name"] = games_data["Name"].str.capitalize()
```

```python
from datetime import datetime
import random

index = random.randint(0, len(games_data))

# Original string
date_string = games_data['Release_Date'][index]

# Convert to date format
date_format = datetime.strptime(date_string, '%b %d, %Y').date()

print("Converted date format:", date_format)
```

```python
games_data['Release_Date'] = pd.to_datetime(games_data['Release_Date'], format='%b %d, %Y', errors='coerce')
```
